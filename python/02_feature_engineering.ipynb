1. Importing Pandas Library

```python
import pandas as pd
```

2. Loading Cleaned Dataset and Checking Dataset

```python
df = pd.read_csv("study_abroad.csv")
```

3. Creating ROI Score Column

```python
df["ROI"] = (df["Average_Graduate_Salary_USD"]/df["Total_Annual_Cost_USD"])
```

```python
df.head()
```

4. Creating Visa Success Score

```python
df["Visa_Success"] = (df["Student_Visa_Approvals"] / df["Student_Visa_Applications"]) * 100
```

```python
df.columns
```

```python
df.head()
```

5. Creating Affordability Score Column

```python
df["Affordability"] = (1/df["Total_Annual_Cost_USD"]) * 100000
```

6. Saving the Dataset After Applying Featured Engineering

```python
df.to_csv("study_abroad_features.csv",index=False)
```
