1. Import Libraries

```python
import pandas as pd
from sqlalchemy import create_engine
```

2. Read CSV

```python
df = pd.read_csv("study_abroad_features_latest.csv", encoding="latin1")
```

3. PostgreSQL Connection

```python
engine = create_engine(
    "postgresql://postgres:admin@localhost:5432/Study_Abroad"
)
```

4. Load into PostgreSQL

```python
df.to_sql(
    "data",          # table name
    engine,
    if_exists="replace",
    index=False
)
```
